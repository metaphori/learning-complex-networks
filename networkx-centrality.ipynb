# Centrality measures

## Betweenneess centrality

```python
import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt
from pathlib import Path
data_dir = Path('.') / 'data'

G = nx.karate_club_graph()
karate_pos = nx.spring_layout(G, k=0.3)
node_color = ['#00aa00' if v%2==0 else '#aa0000' for v in G.nodes]
print(node_color)
nx.draw_networkx(G, karate_pos, node_color=node_color)
plt.show()

degree_c = nx.degree_centrality(G) # dict from nodes to their degree centrality
degree_c = sorted(degree_c.items(), key=lambda item: item[1], reverse=True)
print(degree_c)

betweenness_c = nx.betweenness_centrality(G)
betweenness_c = sorted(betweenness_c.items(), key=lambda item: item[1], reverse=True)
print("Betweenness: ",betweenness_c)

eigenvector_c = nx.eigenvector_centrality(G)
eigenvector_c = sorted(eigenvector_c.items(), key=lambda item: item[1], reverse=True)
print("Eigenvector c.: ", eigenvector_c)

closeness_c = nx.closeness_centrality(G)
closeness_c = sorted(closeness_c.items(), key=lambda item: item[1], reverse=True)
print("Closeness: ", closeness_c)

triangles = nx.triangles(G)
triangles = sorted(triangles.items(), key=lambda x:x[1], reverse=True)
print("Triangles: ", triangles)

clustering = nx.clustering(G)
clustering = sorted(clustering.items(), key=lambda x:x[1], reverse=True)
print("Clustering", clustering)
```
