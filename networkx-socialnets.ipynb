## Global structure

```python
import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt
from pathlib import Path
data_dir = Path('.') / 'data'

G_social = nx.read_edgelist(data_dir / 'mcauley2020-facebook-combined.txt')

G_min_ring = nx.watts_strogatz_graph(16, 2, 0)
G_small_ring = nx.watts_strogatz_graph(16, 4, 0)

plt.figure(figsize=(7.5, 2.75))
plt.subplot(1,2,1)
pos = nx.circular_layout(G_min_ring)
nx.draw_networkx(G_min_ring, pos=pos, with_labels=False)
plt.subplot(1,2,2)
pos = nx.circular_layout(G_small_ring)
nx.draw_networkx(G_small_ring, pos=pos, with_labels=False)
plt.show()

print(nx.average_shortest_path_length(G_min_ring))
print(nx.average_shortest_path_length(G_small_ring))

print(nx.average_clustering(G_min_ring))
print(nx.average_clustering(G_small_ring))

# print(nx.average_shortest_path_length(G_social))
# print(nx.average_clustering(G_social))

# G_ring = nx.watts_strogatz_graph(1000, 10, 0)
# print(nx.average_shortest_path_length(G_ring)) # 200.45011252813202 (200 degrees of sep!)
# print(nx.average_clustering(G_ring)) # 0.6666666666666546 (large mean clustering coeff!)

G_small_random = nx.watts_strogatz_graph(16, 4, 1)
pos = nx.circular_layout(G_small_random)
nx.draw_networkx(G_small_random, pos=pos, with_labels=False)
```
